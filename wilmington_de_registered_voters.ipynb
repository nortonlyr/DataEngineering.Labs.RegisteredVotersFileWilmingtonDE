# Merging Two DataFrames

Using the data from the documentation, create a Pandas DataFrame that contains the party code and the corresponding legal name for each party.

```python
# legal_names = {}

legal_names = {'Party Code': ['A', 'B', 'D', 'E', 'G', 'H', 'I', 'J', 'L', 'M', 'O', 'Q', 'R', 'T', 'V', 'W','Y', 'Z']
              , 'Legal Name': ['PARTY', 'AMERICAN DELTA', 'DEMOCRTIC PARTY', 'LIBERAL', 'NON PARTISAN',
              'GREEN PARTY', 'UNAFFILIATED', 'WORKING FAMILIES PARTY', 'LIBERTARIAN PARTY',
              'REFORM PARTY', 'OTHER / ALL OTHER PARTIES', 'CONSTITUTION PARTY', 'REPUBLICAN PARTY','CONSERVATIVE', 
                'NATURAL LAW PARTY', 'INDEPENDENT PARTY OF DEL','THE BLUE ENIGMA PARTY', 'SOCIALIST WORKERS PARTY']}


```

```python
# df_legal_names = pd.DataFrame(legal_names)
df_legal_names = pd.DataFrame(legal_names)
```

```python
## df_legal_names
df_legal_names
```

Create a Pandas DataFrame that contains the Party code and the corresponding number of registered voters for said party.

```python
# Create DataFrame df_party_counts here 
df_party_counts = pd.DataFrame(grouped_p.size())
```

```python
# df_party_counts 
df_party_counts = df_party_counts.reset_index()
df_party_counts = df_party_counts.rename(columns={0: 'PARTY COUNTS'})
df_party_counts
```

Merge the two DataFrames.

```python
# Merge DataFrame here 
merged_df = pd.merge(df_legal_names, df_party_counts, left_on = 'Party Code', right_on = 'PARTY')
```

```python
# Display new DataFrame here 
merged_df
```

Create a bar graph again but using the legal name for each party as the value on the x axes.

```python
merged_df.plot.bar(x = 'Legal Name', y = 'PARTY COUNTS')
```

```python
#plt.bar method
plt.bar(merged_df['Legal Name'], merged_df['PARTY COUNTS'])
plt.xticks(rotation = 'vertical')
plt.xlabel('Legal Name')
plt.ylabel('PARTY COUNTS')
```
